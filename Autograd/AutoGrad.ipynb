* Autograd is PyTorch's automatic differentiation engine that tracks operations on tensors with requires_grad=True.
* It builds a computational graph dynamically, where each operation is a node, enabling backpropagation.
* Calling .backward() on a scalar loss computes gradients for all tensors involved in its computation.
* Gradients are accessed via .grad and can be stopped using with torch.no_grad() or .detach().

```python
import torch
```

```python
x=torch.tensor(5,requires_grad=True,dtype=torch.float)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
y.backward() #all gradients will be calculated
```

```python
x.grad
```

```python
z=torch.sin(y)
```

```python
x
```

```python
y
```

```python
z
```

```python
z.backward()
```

```python
x.grad
```
